```python
from bs4 import BeautifulSoup
  
# Opening the html file
HTMLFile = open("flights22.html", "r", encoding="utf-8")
  
# Reading the file
index = HTMLFile.read()
  
# Creating a BeautifulSoup object and specifying the parser
Parse = BeautifulSoup(index, 'html')

```

```python
print(index)
```

```python
# znajdowanie numeru lotu, jesli lot nie zaczyna się L to go wywalić, dodać do listy wszystkie numery i policzyć ich liczbę
import regex
listanumerowparser = Parse.find_all('td', {'class' : 'flightsTable-number'})
listanumerow = []
for i in listanumerowparser:
    a = str(i.text)
    if a.startswith("L") == True:
        listanumerow.append(str(i.text))
listanumerow
```

```python
len(listanumerow)
```

```python
lista2=[]
for a in Parse.find_all('a', href=regex.compile("^\/flights\/lo[a-zA-Z0-9]*")):
    lista2.append(str(a['href']))
```

```python
lista2
```

```python
len(lista2)
```

```python
loty22 = []
for i in lista2:
    lista = (i.split("/"))
    lista.pop()
    #usuwa ostatni i pierwszy element listy (były puste)
    lista.pop(0)
    lista.pop(0)
    loty22.append(lista)
loty22
```
